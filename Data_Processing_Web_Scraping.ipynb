```python
import numpy as np # Imports the numpy library
import pandas as pd # Imports the pandas library
from IPython.display import HTML
```

```python
import requests  # Imports the Requests library to handle HTTP requests
```

```python
endpoint = "https://random-data-api.com/api/v2/users"
# As per the website, you can 'DEMO_KEY' as the API key if you are not a heavy user
api_key = "users"

query_params = {"size":100 }

response = requests.get(endpoint, params=query_params)

response
```

```python
data = response.json()
data
```

```python
import time
all_data = []
users_per_request = 100
for i in range(50):  # 50 requests * 100 users/request = 5000 users
    query_params = {"size": users_per_request}
    try:
        response = requests.get(endpoint, params=query_params)
        response.raise_for_status()  # Raise an exception for bad status codes
        data = response.json()
        all_data.extend(data)
        print(f"Fetched users {i*users_per_request+1} to {(i+1)*users_per_request}")
        time.sleep(2) # add a delay to respect API rate limits
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        break  # Exit the loop on error


print(f"Total users fetched: {len(all_data)}")
```

```python
def calculate_age(birthdate):
    today = datetime.date.today()
    age = today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
    return age

# Assuming 'user_data_df' is your DataFrame and 'date_of_birth' is the column name
# containing the date of birth information.  The date format in the column MUST BE
# correctly parsed by pd.to_datetime.

user_data_df['date_of_birth'] = pd.to_datetime(user_data_df['date_of_birth'])

user_data_df['current_age'] = user_data_df['date_of_birth'].apply(calculate_age)

user_data_df
```

### 1.Select a recent news/magazine article related to business analytics that interests you and make a wordcloud using the template provided in the session 4 jupyter notebook. Provide the full text of the article in the notebook so that we can evaluate your work

```python
# Import the BA_text_ai.txt file.
# Keep the BA_text_ai.txt file in the same folder as this notebook
with open("BA_text_ai.txt", 'r', encoding='utf-8') as file:
    BA_text_ai = file.read()
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
# Tokenize the text
tokens = nltk.word_tokenize(BA_text_ai)

# Keep only the tokens that are alphabetic and convert all to lower case
tokens = [word.lower() for word in tokens if word.isalpha()]  # Keep only alphabetic tokens

# Remove stop words
tokens = [word for word in tokens if word not in stopwords.words('english')]
```

```python
# Join tokens to recreate the text as a single string
cleaned_text = ' '.join(tokens)

# Create a word cloud (note the background color)
wordcloud = WordCloud(width=800, height=400, background_color ='pink').generate(cleaned_text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
# Join tokens to recreate the text as a single string
cleaned_text = ' '.join(tokens)

# Create a word cloud (note the background color)
wordcloud = WordCloud(width=800, height=400, background_color ='pink').generate(cleaned_text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide the axes
plt.show()
```

#### The sentiment score for each quote generated by stanza has several numbers, one for each sentence.Create an average score by averaging over all sentences. Create a new column to store the average score

```python
!pip install stanza
import stanza
```

```python
#Downloading the English Model
stanza.download('en')
```

```python
#Let's perform sentiment Analysis for the qoutes file
quotes_subset = quotes.sample(n=1000)
```

```python
quotes_subset.head()
```

```python
#Creating a new column in the dataframe and store the sentiment values
nlp_sent = stanza.Pipeline(lang='en', processors='tokenize,mwt,pos,lemma,sentiment')
def sent_func(quote):
    doc = nlp_sent(quote)
    sent_list = []
    for i, sentence in enumerate(doc.sentences):
        sent_list.append(sentence.sentiment)
    return sent_list

# We will now use this function to create the sent_list for every row
quotes_subset['sentiment'] = quotes_subset['quote'].apply(lambda x: sent_func(x))
```

```python
quotes_subset
```

```python
#Deriving scores from two methods
vader_scores = quotes['scores'].apply(lambda x: x['compound'])
stanza_scores = quotes_subset['scores'].apply(lambda x: x['compound'])
```

```python
vader_scores
```

```python
stanza_scores
```

### Let us examine if the two methods are consistent in scoring the sentiment. Make a scatterplot of VADER compound score against the average stanza score. What would you expect if the two methods are consistent? What do you observe? What is your conclusion?

```python
#Creating a scatter Plot for both the scores
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(vader_scores, stanza_scores, alpha=0.5)  # alpha for transparency
plt.xlabel("VADER Sentiment Scores (quotes)")
plt.ylabel("VADER Sentiment Scores (quotes_subset)")
plt.title("Comparison of Sentiment Scores (VADER vs Stanza)")
plt.grid(True)
plt.show()
```

From the above Scatter Plot it is clear that corelation is more in the words that has  positive values.
